```python

# Read CSV file
def read_csv(file_path):
    return pd.read_csv(file_path)

# Function to calculate statistics for the Price column
def calculate_price_statistics(data):
    price_stats = {}
    price_stats['Min'] = int(data['Price'].min())                   # Minimum as int
    price_stats['Max'] = int(data['Price'].max())                   # Maximum as int
    price_stats['Mean'] = float(data['Price'].mean())               # Mean as float
    price_stats['Median'] = float(data['Price'].median())           # Median as float
    price_stats['Std Dev'] = float(data['Price'].std())             # Standard Deviation as float
    price_stats['Mode'] = int(data['Price'].mode()[0]) if not data['Price'].mode().empty else None  # Mode as int
    return price_stats

# Print statistics in two-column format
def print_statistics_Price(stats):
    #print(f"{'Statistic':<15}{'Value'}")
    print("Price")
    print("-" * 25)
    for key, value in stats.items():
        print(f"{key:<15}{value}")

# Function to calculate statistics for the Price column
def calculate_LandSurface_statistics(data):
    LandSurface = {}
    LandSurface['Min'] = int(data['LandSurface'].min())                   # Minimum as int
    LandSurface['Max'] = int(data['LandSurface'].max())                   # Maximum as int
    LandSurface['Mean'] = float(data['LandSurface'].mean())               # Mean as float
    LandSurface['Median'] = float(data['LandSurface'].median())           # Median as float
    LandSurface['Std Dev'] = float(data['LandSurface'].std())             # Standard Deviation as float
    LandSurface['Mode'] = int(data['LandSurface'].mode()[0]) if not data['LandSurface'].mode().empty else None  # Mode as int
    return LandSurface

# Print statistics in two-column format
def print_statistics_LandSurface(stats):
    #print(f"{'Statistic':<15}{'Value'}")
    print("\nLandSurface")
    print("-" * 25)
    for key, value in stats.items():
        print(f"{key:<15}{value}")

# Function to calculate statistics for the Price column
def calculate_LivingArea_statistics(data):
    LivingArea = {}
    LivingArea['Min'] = int(data['LivingArea'].min())                   # Minimum as int
    LivingArea['Max'] = int(data['LivingArea'].max())                   # Maximum as int
    LivingArea['Mean'] = float(data['LivingArea'].mean())               # Mean as float
    LivingArea['Median'] = float(data['LivingArea'].median())           # Median as float
    LivingArea['Std Dev'] = float(data['LivingArea'].std())             # Standard Deviation as float
    LivingArea['Mode'] = int(data['LivingArea'].mode()[0]) if not data['LivingArea'].mode().empty else None  # Mode as int
    return LivingArea

# Print statistics in two-column format
def print_statistics_LivingArea(stats):
    #print(f"{'Statistic':<15}{'Value'}")
    print("\nLivingArea")
    print("-" * 25)
    for key, value in stats.items():
        print(f"{key:<15}{value}")

# LivingArea, BedroomCount
# Main function
def main():
    data = read_csv(file_path)
    
    # Calculate statistics for Price column
    price_statistics = calculate_price_statistics(data)
    LandSurface_statistics = calculate_LandSurface_statistics(data)
    LivingArea_statistics = calculate_LivingArea_statistics(data)

    # Print statistics in two-column format
    print_statistics_Price(price_statistics)
    print_statistics_LandSurface(LandSurface_statistics)
    print_statistics_LivingArea(LivingArea_statistics)

main()
```

1. Average Price Per Appartment per Municipality:




```python
data = pd.read_csv(file_path)

# 1. Filter the data to include only rows where PropertyType is "APARTMENT"
apartment_data = data[data['PropertyType'] == 'APARTMENT']

# 2. Group by PostalCode and calculate the average Price for apartments in each PostalCode
average_price_per_postalcode = apartment_data.groupby('PostalCode')['Price'].mean().reset_index()

# 3. Rename columns for better readability
average_price_per_postalcode.columns = ['PostalCode', 'AveragePrice']

# 4. Sort the values by PostalCode for better visualization
average_price_per_postalcode = average_price_per_postalcode.sort_values(by='PostalCode')

# 5. Plotting the chart
plt.figure(figsize=(10, 6))
plt.bar(average_price_per_postalcode['PostalCode'].astype(str), average_price_per_postalcode['AveragePrice'], color='skyblue')

# Adding labels and title
plt.xlabel('Postal Code')
plt.ylabel('Average Price (Apartments)')
plt.title('Average Price Per Apartment per Postal Code')

# Rotating the x-ticks for readability
plt.xticks(rotation=90)

# Display the chart
plt.tight_layout()
plt.show()

```

10 most expensive and 10 cheapest postal codes based on the average price for apartments

4. Least/Most Expensive municipalities in Flanders/wallonia/ Belgium in terms of price per m2, average and median price.



average, median, and mean of the price per square meter for each region (Flanders, Wallonia and Brussels)

```python
# Ensure Price and LivingArea are numeric
data['Price'] = pd.to_numeric(data['Price'], errors='coerce')
data['LivingArea'] = pd.to_numeric(data['LivingArea'], errors='coerce')

# Calculate price per square meter (Price / LivingArea)
data['PricePerM2'] = data['Price'] / data['LivingArea']

# Clean data by dropping rows with missing or zero Price or LivingArea
data_cleaned = data.dropna(subset=['Price', 'LivingArea'])
data_cleaned = data_cleaned[data_cleaned['LivingArea'] > 0]

# Function to plot average, median, and mean price per m² for a given region
def plot_stats_price_per_m2(region_name):
    # Filter data for the given region
    region_data = data_cleaned[data_cleaned['Region'] == region_name]

    # Group by PostalCode and calculate mean, median, and average PricePerM2
    region_stats = region_data.groupby('PostalCode').agg(
        AveragePricePerM2=('PricePerM2', 'mean'),
        MedianPricePerM2=('PricePerM2', 'median'),
        StdDevPricePerM2=('PricePerM2', 'std')
    ).reset_index()

    # Plot the stats in a single chart
    plt.figure(figsize=(12, 6))

    # Plot the Average Price Per m²
    plt.plot(region_stats['PostalCode'].astype(str), region_stats['AveragePricePerM2'], label='Average Price Per m²', marker='o')

    # Plot the Median Price Per m²
    plt.plot(region_stats['PostalCode'].astype(str), region_stats['MedianPricePerM2'], label='Median Price Per m²', marker='x')

    # Plot the Standard Deviation Price Per m²
    plt.plot(region_stats['PostalCode'].astype(str), region_stats['StdDevPricePerM2'], label='Standard Deviation of Price Per m²', marker='^')

    # Set chart labels and title
    plt.xlabel('PostalCode')
    plt.ylabel('Price per m²')
    plt.title(f'Price Per m² Statistics in {region_name}')
    plt.xticks(rotation=90)
    plt.legend()

    # Display the chart
    plt.tight_layout()
    plt.show()

# Plot stats (average, median, std deviation) for Flanders
plot_stats_price_per_m2('Flanders')

# Plot stats (average, median, std deviation) for Wallonia
plot_stats_price_per_m2('Wallonia')

# Plot stats (average, median, std deviation) for Brussels
plot_stats_price_per_m2('Brussels')
```

average, median, and mean price per m² for the 10 cheapest and 10 most expensive postal codes per region

```python
# Ensure Price and LivingArea are numeric
data['Price'] = pd.to_numeric(data['Price'], errors='coerce')
data['LivingArea'] = pd.to_numeric(data['LivingArea'], errors='coerce')

# Calculate price per square meter (Price / LivingArea)
data['PricePerM2'] = data['Price'] / data['LivingArea']

# Clean data by dropping rows with missing or zero Price or LivingArea
data_cleaned = data.dropna(subset=['Price', 'LivingArea'])
data_cleaned = data_cleaned[data_cleaned['LivingArea'] > 0]

# Function to plot average, median, and std deviation price per m² for the top 10 cheapest and most expensive postal codes
def plot_top_bottom_price_per_m2(region_name):
    # Filter data for the given region
    region_data = data_cleaned[data_cleaned['Region'] == region_name]

    # Group by PostalCode and calculate mean, median, and std deviation PricePerM2
    region_stats = region_data.groupby('PostalCode').agg(
        AveragePricePerM2=('PricePerM2', 'mean'),
        MedianPricePerM2=('PricePerM2', 'median'),
        StdDevPricePerM2=('PricePerM2', 'std')
    ).reset_index()

    # Sort by AveragePricePerM2
    region_stats_sorted = region_stats.sort_values(by='AveragePricePerM2', ascending=False)

    # Select top 10 expensive and 10 cheapest postal codes
    top_10_expensive = region_stats_sorted.head(10)
    top_10_cheap = region_stats_sorted.tail(10)

    # Concatenate top 10 expensive and top 10 cheap postal codes
    top_bottom_stats = pd.concat([top_10_expensive, top_10_cheap])

    # Plot the stats in a single chart
    plt.figure(figsize=(12, 6))

    # Plot the Average Price Per m² for top and bottom
    plt.plot(top_bottom_stats['PostalCode'].astype(str), top_bottom_stats['AveragePricePerM2'], label='Average Price Per m²', marker='o', color='b')

    # Plot the Median Price Per m² for top and bottom
    plt.plot(top_bottom_stats['PostalCode'].astype(str), top_bottom_stats['MedianPricePerM2'], label='Median Price Per m²', marker='x', color='g')

    # Plot the Standard Deviation Price Per m² for top and bottom
    plt.plot(top_bottom_stats['PostalCode'].astype(str), top_bottom_stats['StdDevPricePerM2'], label='Standard Deviation of Price Per m²', marker='^', color='r')

    # Set chart labels and title
    plt.xlabel('PostalCode')
    plt.ylabel('Price per m²')
    plt.title(f'Price Per m² (10 Cheapest & 10 Most Expensive) in {region_name}')
    plt.xticks(rotation=90)
    plt.legend()

    # Display the chart
    plt.tight_layout()
    plt.show()

# Plot stats for Flanders (10 cheapest & 10 most expensive)
plot_top_bottom_price_per_m2('Flanders')

# Plot stats for Wallonia (10 cheapest & 10 most expensive)
plot_top_bottom_price_per_m2('Wallonia')

# Plot stats for Brussels (10 cheapest & 10 most expensive)
plot_top_bottom_price_per_m2('Brussels')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'C:\\Users\\becod\\AI\\my-projects\\Majid_immoeliza_cleanup\\data_test\\Details\\filtered_data_residential.csv'  # You can provide the actual file path here
data = pd.read_csv(file_path)

# Define custom sorting orders for KitchenType aspects (state, brand, model)
kitchen_state = {
    'NOT_INSTALLED': 0,  
    'SEMI_INSTALLED': 2,          
    'INSTALLED': 3,
    'SEMI_EQUIPPED': 3,
    'EQUIPPED': 4,
    'HYPER_EQUIPPED': 5
}

kitchen_brand = {
    'Belgium': 1,             
    'Germany': 2,
    'USA': 3,
    'China': 1,
}

kitchen_model = {
    'Eggo': 10,
    'Dovy Keukens': 9.5,
    'Van Hoecke': 9,
    'IXINA': 8.5,
    'Siematic': 7,
    'Obumex': 6.5,
    'Nolte Küchen': 6,
    'Eggersmann': 5.5,
    'Belmoba': 5,
    'DSM Keukens': 4.5 
}

# Order for kitchen types
kitchen_order = {
    'China': 1,
    'Belgium': 2,             
    'Germany': 3,
    'USA': 4
}

# Define custom sorting orders for StateBuilding
state_building_order = {
    'TO_RESTORE': 1,
    'TO_RENOVATE': 1,
    'TO_BE_DONE_UP': 2,
    'GOOD': 3,
    'JUST_RENOVATED': 4,
    'AS_NEW': 5
}

# Function to plot the summation of a column (StateBuilding or KitchenType) for each region
def plot_summation_by_region(data, region_name, column_name, title, value_order):
    # Filter data for the given region
    region_data = data[data['Region'] == region_name]
    
    # Group by the column (StateBuilding or KitchenType) and count the occurrences of each value
    value_counts = region_data[column_name].value_counts().reset_index()
    value_counts.columns = [column_name, 'Count']
    
    # Sort the values based on the custom order
    value_counts[column_name] = pd.Categorical(value_counts[column_name], categories=value_order.keys(), ordered=True)
    value_counts = value_counts.sort_values(by=column_name)

    # Plotting the results
    plt.figure(figsize=(8, 6))
    plt.bar(value_counts[column_name].astype(str), value_counts['Count'], color='skyblue')

    # Set chart title and labels
    plt.title(f'{title} in {region_name}')
    plt.xlabel(column_name)
    plt.ylabel(f'Count of {column_name}')
    
    # Rotating the x-axis labels for better readability
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# List of regions to plot
regions = ['Flanders', 'Wallonia', 'Brussels']

# 1. Plotting for StateBuilding for each region with custom order
for region in regions:
    plot_summation_by_region(data, region, 'StateBuilding', 'StateBuilding Distribution', state_building_order)

# 2. Plotting for KitchenType for each region with custom order
for region in regions:
    plot_summation_by_region(data, region, 'KitchenType', 'KitchenType Distribution', kitchen_order)
```
